```python
# calculating the no of  apps having family category
count=0
for i in data["Category"]:
    if(i=="FAMILY"):
        count+=1
print(count)
    
```

```python
# caluculating the no of apps having ART design and lifestyle category
count=0
for i in data["Category"]:
   if(i=="ART_AND_DESIGN" or i=="LIFESTYLE"):
      count=count+1 
print(count)
```
