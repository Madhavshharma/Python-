### handling Missing Values 


Methods for Handling Missing Values
There are several ways to handle missing values:

Removing Data

Removing columns with too many missing values.

Removing rows that contain missing values.

Filling Missing Values (Imputation)

For numerical data: Use mean, median, or mode.

For categorical data: Use mode or introduce a new category.

Advanced Imputation Techniques

Interpolation to estimate missing values.

Predictive models to estimate missing values based on other variables.

### 5. Assigning a New Category for Categorical Data

    
<p>
    Instead of using mode,
    we can introduce a new category like "Not Reviewed".using fillna()
</p>

```python
df6=df5.copy()
df6.head()
```

```python
df6.sample(10)
```

```python
df6["last_review"].fillna("not_reviewd",inplace=True)
df6.isnull().sum()
```

```python
count=0
df6.sample(12)
for i in df6["last_review"]:
    if i=="not_reviewd":
        count=count+1
print(count)
    
```

```python
df6
```

```python
for review in df6["last_review"].unique():
    count=0
    for i in df6["last_review"]:
        if i==review:
            count=count+1
    print(review,count)
```
