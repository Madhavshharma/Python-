```python
import pandas as pd
import numpy as np 
```

```python
data=pd.read_csv("googleplaystore.csv").dropna()
data
```

### Q1. how many apps in each category 

```python
data["Category"].unique()
```

```python
for name in data["Category"].unique():
    ct=0
    for i in data["Category"]:
        if i==name:
            ct=ct+1
    print(name,ct)
```

<p> Saving the result in the dictionary 

```python
categorys={}
for name in data["Category"].unique():
    ct=0
    for i in data["Category"]:
        if i==name:
            ct=ct+1
    categorys[name]=ct
```

```python
categorys[name]=ct
categorys
```

```python
print(categorys)
```
